```python
import string
f=open("test.txt","r")
txt=f.read()
l_txt=str.lower(txt)
l_txt_rp =l_txt.replace("\n"," ")
l_txt_rp = l_txt_rp.translate(str.maketrans('', '', string.punctuation)) 
data=l_txt_rp.split(" ")
unique_data=sorted(list(set(data)))
for x in range(len(unique_data)):
    i=0
    for y in range(len(data)):
        if data[y]==unique_data[x]:
            i=i+1
        else:
            continue
    print(unique_data[x], ":", i)
```
